```python
import pandas as pd
import numpy as np
import seaborn as sns
from google.colab import files
import matplotlib.pyplot as plt

from sklearn.preprocessing import PowerTransformer
```

```python
df=pd.read_excel('/content/drive/MyDrive/skripsi/2021.xlsx')
```

```python
df
```

```python
data = df[['elevasi', 'slope', 'ndbi', 'ndvi', 'ndwi','no','co','ntl','wave','arus','pop','gp']]
```

```python
trans_yeo_johnson = PowerTransformer(method="yeo-johnson")
data_trans = trans_yeo_johnson.fit_transform(data)
```

```python
data_trans_df = pd.DataFrame(data_trans, columns=['elevasi', 'slope', 'ndbi', 'ndvi', 'ndwi','no','co','ntl','wave','arus','pop','gp'])
data_trans_df
```

```python
data_trans_df.to_excel("2023_transformfix.xlsx", index=False)
files.download("2023_transformfix.xlsx")
```

```python
import scipy.stats as stats

# Membuat histogram untuk setiap kolom
data_trans_df.hist(bins=30, figsize=(12, 8), grid=False)
plt.suptitle("Histogram dari Data yang Ditransformasi (Yeo-Johnson)", fontsize=14)
plt.show()

# Membuat QQ Plot untuk setiap kolom
fig, axes = plt.subplots(3, 3, figsize=(12, 12))  # 3x3 grid plot
fig.suptitle("QQ Plot dari Data yang Ditransformasi (Yeo-Johnson)", fontsize=14)

columns = data_trans_df.columns
for i, ax in enumerate(axes.flatten()):
    if i < len(columns):
        stats.probplot(data_trans_df[columns[i]], dist="norm", plot=ax)
        ax.set_title(columns[i])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
